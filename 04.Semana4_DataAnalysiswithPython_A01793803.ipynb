#### **PAQUETES DE PYTHON**

1. Científico

* Pandas: Manipulación de datos como tipos tablas (*dataframes*)
* Numpy: Para operaciones con *arrays* o matrices
* SciPy: Resulve problemas matemáticos como derivadas, integrales, etc.

2. Visualización

* Matplotlib: Gráficos populares, box plot, es la más usadas
* Seaborn: Robusta que pueda realizar hasta series temporales

3. Algoritmos (método de aprendizaje automático)

* Scikit-learn: *machine learning*, clasificación, regresión, segmentación
