## Analysis of Variance ANOVA

- Comparasión estadística por grupos
- Ejemplo: Promedio de precio para diferentes vehículos

**ANOVA** 

- Encuentra las correlaciones entre diferentes grupos de variables categóricas
- F-test score: Variación entre la media de grupos dividido por la variación dentro del mismo grupo
- p-value: El grado de confianza
- Valores pequeños de F significa pobre correlacion entre las variables categóricas y su variable objetivo
- Valores grandes de F significa alta correlacion entre las variables categóricas y su variable objetivo

## Correlation

Es una métrica estadística para medida que tan diferentes son las variables interpendientes
- La correlación no necesariamente implica causalidad

`sns.regplot()`

## Correlation statistics

**Correlación de Pearson**

- Mide que tan fuerte es la relación entre dos variables:
    - Coeficiente de correlación
    - P-value

**Coeficiente de correlación**

- +1 --> Relación positiva
- -1 --> Relación negativa
- 0 --> No hay relación

**P-value**

- < 0.001 Fuerte certeza en el resultado
- < 0.05 Moderada certeza en el resultado
- < 0.1 Debil certeza en el resultado
- '> 0.1 No hay certeza en el resultado

`stats.pearsonr()`

## ANOVA

```python
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
grouped_test2.get_group('4wd')['price']
```

```python
# ANOVA
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val) 
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('rwd')['price'])  
   
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('fwd')['price'])  
 
print("ANOVA results: F=", f_val, ", P =", p_val)
```
