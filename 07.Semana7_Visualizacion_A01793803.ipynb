### **3. Limpia los datos eliminando los registros nulos o rellena con la media de la columna**

```python
row_target = pd.isnull(df['Class'])
df_target = df[row_target]
df_v1 = df[~row_target].reset_index(drop=True)
```

#### ***Variables categóricas***

Teniendo en cuenta que a pesar que las variables `Gender`, `Marital_status`, `Education` y `Age`, contiene valores numéricos, estas pueden ser consideradas categóricas. A partir de esto, comenzaremos el análisis de `Gender`. Verificamos la cantidad de datos total que contiene el conjunto y buscaremos el valor que hace falta para esta variable

```python
row_gender = pd.isnull(df_v1['Gender'])
df_gender = df_v1[row_gender]
```

```python
df_v2 = df_v1[~row_gender].reset_index(drop=True)
```

Procedemos a verficar las otras columnas categóricas mencionadas arriba `Matiral status` y `Education`

```python
row_cat = pd.isnull(df_v2['Marital status'])
df_cat = df_v2[row_cat]
df_v2 = df_v2[~row_cat].reset_index(drop=True)
row_cat2 = pd.isnull(df_v2['Age'])
df_cat2 = df_v2[row_cat2]
```

Verificamos que solo son tres observaciones con valores nulos. Existen muchas formas de decisión a la hora de imputar este tipo de valores. Acá, se realizará una de ellas y es usar alguna de las variables que se infiera tiene un comportamiento que pueda representar aquella que poseen valores nulos. La alternativa rápida es borrar esos registros ya que solo representarian el 0.01% de las muestras, pero se decide optar por esta estrategía para poder aplicar en otras variables. Se tomará como base la variable `Amount of the given credit` y se marcaran rangos que puedan representar los valores de la `Age`. Primero se realiza un histograma para evidencia la distribución de esta y luego se procede a encontrar estos rangos.

```python
datos_no_nulos = df_v2[~row_cat2].reset_index(drop=True)
datos_nonulos = datos_no_nulos[['Amount of the given credit', 'Age']]
agrupamiento_nonulos = datos_nonulos.groupby(pd.cut(datos_nonulos['Amount of the given credit'], np.arange(int(datos_nonulos['Amount of the given credit'].min())\
    ,int(datos_nonulos['Amount of the given credit'].max())+10000, int(datos_nonulos['Amount of the given credit'].min())))).median()
agrupamiento_nonulos.head()
```

```python
df_cat2 = df_cat2.reset_index(drop=True)
datos_imp = agrupamiento_nonulos.reset_index(drop=True)
row = 0
while row < df_cat2.shape[0]:
    for i in range(datos_imp.shape[0]-1):
        
        if df_cat2.loc[row, 'Amount of the given credit'] > datos_imp.loc[i, 'Amount of the given credit'] \
            and df_cat2.loc[row, 'Amount of the given credit'] <= datos_imp.loc[i+1, 'Amount of the given credit']:
            df_cat2.loc[row, 'Age'] = datos_imp.loc[i, 'Age']
    
    row = row+1
```

```python
data_final = pd.concat([df_cat2, datos_no_nulos], axis=0)
```

### **7. Reduce las dimensiones con PCA, si consideras necesario.**
* Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%
* Indica la importancia de las variables en cada componente

Considerando que las variables `ID`, `Amount of the given credit`, `Gender`, `Education`, `Marital Status` y `Age` contienen información particular de los clientes, describiendo especificamente a cada uno de estos. Se realizará la reducción de dimensionalidad a las otras variables como se muestra a continuación.

```python
col1_filter = [col for col in data_final.columns if col.startswith('Amount paid')]
col2_filter = [col for col in data_final.columns if col.startswith('Amount of bill')]
col3_filter = [col for col in data_final.columns if col.startswith('the repayment')]
```

Se muestra un gráfico de correlación para mostrar la primera aproximación de `PCA` y es ver cuales están correlacionadas, las cuales posiblemente podrán reducirse a menos componentes:

```python
fig, ax = plt.subplots(figsize=(12,10))
final_all = data_final[col1_filter+col2_filter+col3_filter]
final_all_status = final_all.corr()
sns.heatmap(final_all_status, annot=True, cmap='BuGn', fmt='.1g', ax=ax)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
col_total = col1_filter + col2_filter + col3_filter
list_col = []
for col in col_total:
    var = df_scal[col]
    list_col.append(var)
    
# COVARIANZA
# =================================
var_total = np.cov(list_col)
var_col = df_scal
total = var_col.var().sum()

list_var = []
for row in range(len(col_total)):
    var = var_total[row][row]
    list_var.append(var)
```

```python
print('Varianza total \t\t\t:', total.round(2) )

for i in range(len(col_total)):
    print(f'Varienza prop. {col_total[i]} \t:', ((list_var[i]/total)*100).round(2),'%' )
```

```python
pca_total = PCA(n_components=9)
pca_total.fit(df_scal[col1_filter+col2_filter+col3_filter])

pcaSummary = pd.DataFrame({'Desviación estandar': np.sqrt(pca_total.explained_variance_),'Proporción de la varianza': pca_total.explained_variance_ratio_,\
                          'Proporción acumulada': np.cumsum(pca_total.explained_variance_ratio_)})
pcaSummary = pcaSummary.transpose()
pcaSummary.columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5','PC6','PC7','PC8', 'PC9']
pcaSummary = pcaSummary.round(3)
pcaSummary
```

Se evidencia que con 2 componentes se logra representar el `72.9%` de la varianza de las 18 variables a considerar para reducción de dimensionalidad. El `90%` se logra con 5 componentes.

### Importancia de variables

```python
pca_total.components_.transpose().shape
```

```python
pca_componentes = pd.DataFrame(pca_total.components_.transpose(), 
                                columns = pcaSummary.columns,
                                index = df_scal[col1_filter+col2_filter+col3_filter].columns
                                )
pca_componentes
```

* La componente que muestra la explicación del `70%` de la varianza es `PC2`, y como variable más importante es el pago de mayo de 2005 (`Amount paid in May, 2005`).
* Para el caso del `90%` de la varianza, la variable más importante de la componente `PC5` es: `the repayment status in August`. A pesasr de que el valor es negativo, la importancia es el valor absoluto, el signo muestra la dirección de la variable en el vector
