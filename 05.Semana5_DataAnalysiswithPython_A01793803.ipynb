## **1. Preprocessing data in python**

El proceso de preprocesamiento trata de mapear cada una de las observaciones o filas de nuestro conjunto de datos para transformar los formatos con el objetivo de prepararlo para un análisis de datos --> `Data cleaning or data wrangling`

El proceso de `limpieza de datos` se puede clasificar de la siguiente manera:
    
1. Identificar y tratar los valores nulos
2. Formateo de los datos
3. Normalización de los datos (Escalado)
4. Codificación de datos [Data Binning]
5. Tunear variables categóricas a numéricas    

## **2. Dealing with missing values in Python**

* Un valor nulo ocurre cuando no existe datos (valor) guardado para las variables [en cada una de las observaciones]
* Este puede estar representado por `?`, `N/A`, `0` o solo celdas vacías

### Soluciones
* Revisar la fuente del conjunto de datos original
* Eliminar los valores nulos
    * Borrar la variable
    * Borrar la observación
* Reemplazar el valor nulo
    * Reemplazar por el promedio (puntos similares)
    * Reemplazar por la frencuencia
    * Reemplazar basados en otras funciones
* O dejar el valor nulo

Usar `dataframes.dropna()`

*Ejemplo*:

`df.dropna(subset=["price"], axis=0, inplace=True)`

Usar `dataframe.replace(missing_value, new_value)`

*Ejemplo*:

`mean = df["normalized-losses"].mean()`
`df["normalized-losses"].replace(np.nan, mean)`


## **3. Data Formatting in Python**

*Ejemplos* 

`df["city-mpg"] = 235/df["city-mpg"]
`df.rename(columns={"city_mpg": "city-L/100km"}, inplace=True}` 

Algunas veces se asigna un tipo de datos incorrecto a la característica completa

* Objects
* int64
* Float64

Los tipos de datos pueden ser identificados con `dataframe.dtypes()`

Para convertir a un tipo de datos: `dataframe.astype()`

## **4. Data Normalization in Python**

La normalización es importante para que se puedan considerar cada una de las características con las mismas dimensiones

#### Existen diferentes métodos

* Escalado simple
* Máximo y mínimo
* Escalado estándar

## **5. Binning in Python**

* Binning: Agrupar los valores en `bins`
* Convertir los valores categóricos a numéricos
* Conjuntos de datos numéricos a conjunntos de `bins`

## **6. Turning categorical variables into quantitative variables in Python**

* Muchos modelos estadísticos no pueden tomar variables tipo `object` para los modelos

Una de las soluciones es asignar variables dummies como única categoría

Usar en pandas `pandas.get_dummies()`

## **EJEMPLOS**

```python
import pandas as pd
import numpy as np
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(path, names=headers)
```

```python
df.head()
```

### Missing values

Reemplazar todos los valores nulos que contienen `?` por valores `nan`

```python
df.replace("?", np.nan, inplace=True)
df.head()
```

Se extraen las filas de todos los valores nulos que existen (observaciones)

```python
df_missing = df.isnull()
df_missing.head()
```

Se verifican cuales son las columnas que tienen valores nulos

```python
for column in df_missing.columns.values.tolist():
    print(column)
    print (df_missing[column].value_counts())
    print("") 
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
avg_bore=df['bore'].astype('float').mean(axis=0)
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

## **Question 1**

Based on the example above, replace NaN in "stroke" column with the mean value

```python
df.columns
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0)
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
df.head()
```

se realiza el mismo procedimiento para las otras columnas 

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)

avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
df["num-of-doors"].replace(np.nan, "four", inplace=True)
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

### Formateo de los datos

Se cambian algunos formatos para tener un conjunto de datos limpio

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

### Estandarización de los datos

```python
df['city-L/100km'] = 235/df["city-mpg"]
```

## **Question 2**

According to the example above, transform mpg to L/100km in the column of "highway-mpg" and change the name of column to "highway-L/100km"

```python
df['highway-L/100km'] = 235/df['highway-mpg']
df = df.drop(columns='highway-mpg')
df.head()
```

### Normalización de los datos

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

## **Question 3**

According to the example above, normalize the column "height"

```python
df['height'] = df['height']/df['height'].max()
```

### Binning

Covertir valores categóricos a valores numéricos

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

### Dummy variables

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
df = pd.concat([df, dummy_variable_1], axis=1)
df.drop("fuel-type", axis = 1, inplace=True)
```
